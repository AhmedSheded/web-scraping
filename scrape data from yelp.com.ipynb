```python
from selenium import webdriver
import pandas as pd
import time
```

```python
driver = webdriver.Chrome(executable_path= '/usr/bin/chromedriver')
yelp = 'https://www.yelp.com/search?find_desc=Dentists&find_loc=San+Francisco%2C+CA&ns=1'
driver.get(yelp)
driver.maximize_window()
time.sleep(2)
```

# Target The Data

```python
# names 
names = driver.find_elements_by_xpath('//span[@class=" css-1pxmz4g"]/a')

# expirtise
expirtise = driver.find_elements_by_xpath('//div[@class=" border-color--default__09f24__3Epto"]/div/div/div/p')

# stars
stars = driver.find_elements_by_xpath('//div[@class=" attribute__09f24__1La1D display--inline-block__09f24__3SvIn margin-r1__09f24__3PebR border-color--default__09f24__3Epto"]/span/div')

# rating count
rating_count = driver.find_elements_by_xpath('//span[@class="reviewCount__09f24__3GsGY css-e81eai"]')
```

```python
# i =1 
# for n in expirtise:
#     print(i, " ", n.text)
#     i+=1
# print(len(expirtise))

for i in names:
    print(i.text)
```

```python
names_list = []
expirtise_list = []
stars_list = []
rating_list = []
```

```python
for n in names:
    names_list.append(n.text)
    
for e in expirtise:
    expirtise_list.append(e.text)
    
for s in stars:
    stars_list.append(s.get_attribute('aria-label'))
    
for r in rating_count:
    rating_list.append(r.text)
```

```python
stars_list
```

```python
dent_dictionary = { 'Names': names_list, 'Expirtise': expirtise_list, 'Stars': stars_list, 'Rating count': rating_list }
```

```python
dent_dictionary
```

```python
df = pd.DataFrame.from_dict(dent_dictionary)
```

```python
df
```

```python
df['Stars']=df['Stars'].apply(lambda x: x.replace(' star rating', ''))
```

```python
csv = df.to_csv('data.csv', index=False)
excel = df.to_excel('data.xlsx', index=False)
```
